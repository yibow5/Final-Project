Data visualization is a form of information communication that depicts  complex information in graphical form, with the final visuals designed to simplify the data and use it to help users to make decisions. At the same time, data visualization can represent different types and sizes of data, from a few data points to large  datasets. It is everywhere in our lives!

Fisrt things first, the quality of data matters! Before the visualization, we needs to examine the accuracy, clarity, and completeness of the data. If the data is incomplete or inaccurate, our visualization will be faulty as well. A thorough evaluation of data is required before proceeding to the next step. 

Then, we need to choose an appropriate type of visualization. Data visualization can be represented in different forms: charts, tables, graphs, maps, dashbpards... Charts are a common way to represent data because they depict different types of data and allow for data comparison, exploration, etc. The decision to use a chart type depends on two main factors: the data to be communicated and the context. 

### Line Chart   
A line chart is mainly used to show the change of data over a period of time. The x axis tpyically indicates the date, and the y axis shows the value. Line graphs are useful to show data variables and trends very clearly and can help to make predictions about the results of data. They can also be used to display several dependent variables against one independent variable.   
Allow us to demostrate!

```python
checkbox_samsung = widgets.Checkbox(value=True,description='Show Samsung',disabled=False,indent=False)
checkbox_apple = widgets.Checkbox(value=True,description='Show Apple',disabled=False,indent=False)
checkbox_nokia = widgets.Checkbox(value=True,description='Show Nokia',disabled=False,indent=False)
checkbox_huawei = widgets.Checkbox(value=True,description='Show Huawei',disabled=False,indent=False)
checkbox_xiaomi = widgets.Checkbox(value=True,description='Show Xiaomi',disabled=False,indent=False)
checkbox_lg = widgets.Checkbox(value=True,description='Show LG',disabled=False,indent=False)
checkbox_oppo = widgets.Checkbox(value=True,description='Show Oppo',disabled=False,indent=False)
checkbox_sony = widgets.Checkbox(value=True,description='Show Sony',disabled=False,indent=False)
checkbox_motorola = widgets.Checkbox(value=True,description='Show Motorola',disabled=False,indent=False)
checkbox_htc = widgets.Checkbox(value=True,description='Show HTC',disabled=False,indent=False)

# Create a checkbox for each mobile vendor. 

diplay_checkboxes = widgets.HBox([checkbox_samsung,checkbox_apple,checkbox_nokia,checkbox_huawei,checkbox_xiaomi,
         checkbox_lg,checkbox_oppo,checkbox_sony,checkbox_motorola,checkbox_htc])
#Create a Hbox for checkboxes to display horizontally.
```

```python
def plot(checkbox_samsung,checkbox_apple,checkbox_nokia,checkbox_huawei,checkbox_xiaomi,
         checkbox_lg,checkbox_oppo,checkbox_sony,checkbox_motorola,checkbox_htc):
    fig, ax = plt.subplots(figsize=(20,8))
    if checkbox_samsung:  
        plt.plot(date,samsung,label = "Samsung", color = 'blue')
    if checkbox_apple:
        plt.plot(date,apple,label = "Apple", color = 'green')
    if checkbox_nokia:
        plt.plot(date,nokia,label = "Nokia", color = 'red')
    if checkbox_huawei:
        plt.plot(date,huawei,label = "Huawei", color = 'orange' )
    if checkbox_xiaomi:
        plt.plot(date,xiaomi,label = "Xiaomi", color = 'black')
    if checkbox_lg:
        plt.plot(date,lg,label = "LG", color = 'lime' )
    if checkbox_oppo:
        plt.plot(date,oppo,label = "Oppo", color = 'purple' )
    if checkbox_sony:
        plt.plot(date,sony,label = "Sony", color = 'teal')
    if checkbox_motorola:
        plt.plot(date,motorola,label = "Motorola", color = 'darkolivegreen')
    if checkbox_htc:
        plt.plot(date,htc,label = "HTC", color = 'pink')
    else:
        plt.plot(date,samsung, color = 'blue', visible = False)
        plt.plot(date,apple, color = 'green', visible = False)
        plt.plot(date,nokia, color = 'red', visible = False)
        plt.plot(date,huawei, color = 'orange', visible = False )
        plt.plot(date,xiaomi, color = 'black', visible = False)
        plt.plot(date,lg, color = 'lime' , visible = False)
        plt.plot(date,oppo, color = 'purple' , visible = False)
        plt.plot(date,sony, color = 'teal', visible = False)
        plt.plot(date,motorola, color = 'darkolivegreen', visible = False)
        plt.plot(date,htc, color = 'pink', visible = False)
        
        
    plt.xlabel('Year')
    plt.ylabel('Market Share')
    plt.legend(loc='upper right')
    plt.show()

display_output = widgets.interactive_output(plot, {'checkbox_samsung': checkbox_samsung, 'checkbox_apple': checkbox_apple, 
                                        'checkbox_nokia': checkbox_nokia, 'checkbox_huawei':checkbox_huawei,
                                        'checkbox_xiaomi':checkbox_xiaomi, 'checkbox_lg':checkbox_lg,
                                        'checkbox_oppo':checkbox_oppo, 'checkbox_sony':checkbox_sony,
                                        'checkbox_motorola':checkbox_motorola, 'checkbox_htc':checkbox_htc})


display(diplay_checkboxes,display_output)

# This cell creates interactivity between checkboxes and the line plot. 
```

This multiple line graph idicates the top 10 mobile vendors'market share from 2010 to 2022. By default this graph is empty. By clicking on a specific checkbox, a line plot will apppear. In total of 10 line plots can appear on this figure. This visualization can be used to see the change of a mobile vendor's market share over the years,or compare market share between one or several mobile vendors.

It is a statistical report graph that expresses data with the length of a rectangle, it represented by a series of vertical graphs of varying heights to represent the distribution of data. It is suitable for displaying two-dimensional data sets and displaying the distribution of data. One axis represents the classification dimension that needs to be compared, and the other axis represents the corresponding value.Similar to a pie chart, it allows uers to estimate all values, just at a glance! ### Scatter Plot

Last but not least, scatter plot. It is a great way to display non-linear data. A scatter plot can display all data points, including exact minimum and maximum. Also, the pattern of a scatter plot is easy to observe.

```python
pop = pd.read_csv("https://raw.githubusercontent.com/changli97/visualization/main/population_usa.csv")
# pop = pd.read_csv("/Users/lichang/Desktop/445/population_usa.csv")
pop
```

##### I found this chart from kaggle, which is a dataset website.
##### This chart introduced the trend of population and other attributes changed in US.

```python
scatter = pop.plot(x = 'Fertility Rate', y = 'World Population', figsize=(10,6), rot=90, kind='scatter')
```

I put the Fertility rate as x axis and World Population as y axis, then it shows us a visualization about how Fertility Rate had been affected by the increase number of World Population. In the beginning, the fertility rate is very high when the number of world population is not so much, but the rate began decrease after that. Now let's throw in some interactions. 

```python

pop.columns
categorical_column = [column for column in pop.columns]
categorical_column
from ipywidgets import interact
popx = widgets.Dropdown(
    options = categorical_column,
    value = categorical_column[6],
    description = 'x axis'
)
popy = widgets.Dropdown(
    options = categorical_column,
    value = categorical_column[0],
    description = 'y axis'
)
ui = widgets.HBox([popx,popy])

def relplot(column_x, column_y):
    p = sns.relplot(data = pop, x=column_x, y=column_y)

export = widgets.interactive_output (relplot, {'column_x' : popx, 'column_y': popy})
display(ui, export)
```

Here is a interactive that we can choose the X and Y axis to get the different visulizations, we can also choose the Year as Y axis to find out the relationship between Year and Fertility rate, which is similar to the upper picture since the world population and Year are positive correlation
